# Pandas Built-in Data Visualization

```python
import numpy as np
import seaborn as sns
import pandas as pd
%matplotlib inline
df1 = pd.read_csv('df1',index_col=0)
df1.head()
```

```python
df2=pd.read_csv('df2')
df2.head()
```

```python
df1['A'].hist(bins=30)
```

```python
df1['A'].plot(kind='hist', bins=30)
```

```python
df1['A'].plot.hist()
```

```python
df2.plot.area(alpha=0.4)
```

```python
df2.plot.bar(stacked=True)
```

```python
df1['A'].plot.hist(bins=50)
```

```python
df1.plot.line(y='B',figsize=(12,3),lw=1)
```

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm') # Datapoints are colored relative to C column
```

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*50) # Datapoints are sized relative to C column
```

```python
df1.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])
```

```python
df.plot.hexbin(x='a',y='b', gridsize=25,cmap='coolwarm')
```

```python
df['a'].plot.kde()
```

```python
df['a'].plot.density()
```
