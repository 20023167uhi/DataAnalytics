# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional method to plot
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Orientated Method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8,])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
# Figure a blank canvas to inroduce plots to using the "add_axes". The 4 Values in "add_axes" are between 0 and 1.
# add_axis[origin x%, origin y%, size x%, sizey%]

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

#axes.plot(x,y)
#for current_ax in axes:
#   current_ax.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout() # Takes care of overlapping plots
```

```python
fig = plt.figure(figsize=(8,2)) # Use dpi to declare resolution

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2)) # Use dpi to declare resolution

axes[0].plot(x,y)
axes[1].plot(x,y)

plt.tight_layout() # Takes care of overlapping plots
```

```python
fig.savefig('my_picture.png', dpi=200)
```

# Figure Size and DPI

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')

ax.set_title('Title')
ax.set_ylabel('Y Label')
ax.set_xlabel('X Label')

ax.legend(loc=5)
```

# Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF00FF', 
        linewidth=2, 
        alpha=0.5, 
        linestyle='-',
        marker='o', 
        markersize=20,
        markerfacecolor='Yellow',
        markeredgewidth=3,
        markeredgecolor='Green',
       ) 
# Color can use RGB Hex values or string values such as 'Orange', 'Blue', 'Green'...
# linewidth, or 'lw', width of line
# Alpha = Transparency
# Marker - plot point markers: '+','*','o','1','2','3'
# Markersize size of marks
```

```python
ax.plot(x,y,color='purple', lw=2, linestyle='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
fig
```
